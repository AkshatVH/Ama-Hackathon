```python
# Define an empty dictionary to store the request cookies
cookie = {}

def getAmazonSearch(search_query):
    # Build the URL for the Amazon search page by concatenating the search query to the base URL
    url = "https://www.amazon.in/s?k=" + search_query
    
    # Print the URL for debugging purposes
    print(url)

    # Send a GET request to the URL, passing in the cookies and header information
    page = requests.get(url, cookies=cookie, headers=header)

    # Check the status code of the request to determine if it was successful
    if page.status_code == 200:
        # If the request was successful, return the response
        return page
    else:
        # If the request was not successful, return an error message
        return "Error"
```

```python
def Searchasin(asin):
    # Concatenate the ASIN and the Amazon URL
    url="https://www.amazon.in/dp/"+asin
    
    # Print the URL for debugging purposes
    print(url)
    
    # Send a GET request to the URL using the requests library
    page=requests.get(url,cookies=cookie,headers=header)
    
    # Check if the request was successful (status code 200)
    if page.status_code==200:
        # Return the page content if successful
        return page
    else:
        # Return an error message if the request was unsuccessful
        return "Error"
```

```python
def Searchreviews(review_link):
    # Define the base URL and concatenate the review link
    url = "https://www.amazon.in" + review_link
    
    # Print the full URL
    print(url)
    
    # Make a GET request to the URL using the requests library
    page = requests.get(url, cookies=cookie, headers=header)
    
    # Check if the request was successful (status code 200)
    if page.status_code == 200:
        # Return the page if the request was successful
        return page
    else:
        # Return an error message if the request was unsuccessful
        return "Error"
```

```python
product_names=[] # Initialize an empty list to store product names

response = getAmazonSearch('Nike+shoes') # Get the response from the Amazon search for "Nike shoes"

soup = BeautifulSoup(response.content) # Create a BeautifulSoup object from the response content

# Find all span tags with the specified class
for i in soup.findAll("span", {'class':'a-size-base-plus a-color-base a-text-normal'}):
    product_names.append(i.text) # Add the product name (i.text) to the product_names list
```

```python
# Initialize an empty list to store the ASINs
data_asin = []

# Make a request to the Amazon search API for Nike shoes
response = getAmazonSearch('Nike+shoes')

# Create a BeautifulSoup object from the response content
soup = BeautifulSoup(response.content)

# Loop through all the HTML elements with the specified class
for i in soup.findAll("div", {'class': "a-section a-spacing-none _bGlmZ_container_3q4Jr _bGlmZ_block_1vI8- _bGlmZ_hFull_2lnNw _bGlmZ_wFull_3f8b2 _bGlmZ_col_358pf"}):
    # Extract the 'data-asin' attribute value and append it to the data_asin list
    data_asin.append(i['data-asin'])
```

```python
#display asin id
data_asin
```

```python
# Initialize an empty list to store the links
link = []

# Loop over each element in the data_asin list
for i in range(len(data_asin)):
    # Call the Searchasin function and store the response
    response = Searchasin(data_asin[i])
    # Use BeautifulSoup to parse the HTML content of the response
    soup = BeautifulSoup(response.content)
    # Find all elements in the HTML with the attribute 'data-hook' equal to "see-all-reviews-link-foot"
    for i in soup.findAll("a", {'data-hook': "see-all-reviews-link-foot"}):
        # Append the 'href' attribute of the element to the 'link' list
        link.append(i['href'])
```

```python
reviews = [] # Initialize an empty list to store the reviews

# Loop over the length of the "link" list
for j in range(len(link)):
    for k in range(100):
        # Make a request to search reviews using the given link and page number
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        
        # Parse the response content using BeautifulSoup library
        soup = BeautifulSoup(response.content)
        
        # Find all HTML elements with the tag "span" and attribute "data-hook" with value "review-body"
        for i in soup.findAll("span", {'data-hook': "review-body"}):
            # Append the text of each found element to the "reviews" list
            reviews.append(i.text)
```
